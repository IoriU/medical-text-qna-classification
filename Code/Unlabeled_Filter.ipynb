```python
def count_kw(s):
  total = 0
  for i in kw:
    if i in s:
      total += 1
  return total
```

```python
count_kw("ablasi retina abses abortus imminens")
```

```python
ranking = []
for i, j in df.iterrows():
  ranking.append((i, count_kw(j.ALL)))
ranking.sort(key=lambda x: x[1], reverse=True)
print(ranking)
```

```python
ranking_nz = [x for x in ranking if x[1] > 2]
print(len(ranking_nz))
print(ranking_nz)
```
